```python
#Importing Libaries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
#create df where job Country is USA
df_usa = df[df["job_country"]=="United States"].copy() 

#creates job_posted_month where month is from job_posted_date
df_usa["job_posted_month"] = df_usa["job_posted_date"].dt.strftime("%B") 

#creates pivot table from df.
df_usa_pivot = df_usa.pivot_table(
index="job_posted_month",
columns="job_title_short", 
aggfunc = "size"
)

#reset index 
df_usa_pivot.reset_index(inplace=True)
#creates month_no from job_posted_month (only the month in a number)
df_usa_pivot["Month_no"] = pd.to_datetime(df_usa_pivot["job_posted_month"], format="%B").dt.month
#sort by month from 1-12
df_usa_pivot.sort_values("Month_no", ascending=True, inplace=True)
#set index from job_posted_momth
df_usa_pivot.set_index("job_posted_month", inplace=True)
#drops month_no 
df_usa_pivot.drop(columns="Month_no", inplace=True)

df_usa_pivot
```

```python
df_usa_software_pivot = pd.read_csv(r"C:\Users\Elmorion\Documents\.Data Science\Python\Python_Data_Project\2_Advanced\software_jobs.csv", index_col="job_posted_month")
df_usa_software_pivot
```

```python
df_usa_merged = df_usa_pivot.merge(df_usa_software_pivot, on="job_posted_month")
```

```python
top_5 = (
 df_usa_merged
.sum()
.sort_values(ascending=False)
.head()
.index
.to_list()
)

top_5
```

```python
df_usa_merged[top_5].plot(kind="line")
plt.title("Monthly Job Postings for Top Tech Jobs in the USA")
plt.xlabel("2023")
plt.ylabel("Job Count")
plt.ylim(0,20000)
plt.legend()
plt.show()
```
